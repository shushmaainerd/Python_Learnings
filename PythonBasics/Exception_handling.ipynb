```python
x=input("Enter a num1")
y=input("Enter a num2")
try:
    z=int(x)/int(y)
except Exception as e:
    print('exception occured:',e)
    z=None
print("Division is :",z)
```

```python
x=input("Enter a num1")
y=input("Enter a num2")
try:
    z=int(x)/int(y)
except ZeroDivisionError as e:
    print('Division of Zero Exception')
    z=None
print("Division is :",z)
```
