```python
#(10,2000)->(20,3000)->(30,4000)->(40,None)
#   1000       2000      3000       4000
#    Node1     Node2      Node3      Node 4
```

```python
# to begin with linked list first we have to create a Node
```

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None

#Creating Nodes

node1 =node(10)  #[10,None] --> stored at address 1000
node2=node(20)   #[20,None] --> stored at address 2000
node3=node(30)   #[30,None] --> stored at address 3000
node4=node(40)   #[40,None] --> stored at address 4000

#I have just creating nodes but there is no relation. so we have to connect the nodes
#Connecting nodes to form a linkedlist
node1.next=node2 #-->2000
node2.next=node3 #-->3000
node3.next=node4 #-->4000

#(10,2000)->(20,3000)->(30,4000)->(40,None)
#   1000       2000      3000       4000

#printing the linked list
current=node1 #node1 address is assigned to current
while current is not None:
    print(current.data, end="->")
    current=current.next
print("None")
```

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None

node1 =node(10) 
node2=node(20)  
node3=node(30)  
node4=node(40)   

node1.next=node2 #-->2000
node2.next=node3 #-->3000
node3.next=node4 #-->4000

head = node1
current=head
while current is not None:
    print(current.data, end="->")
    current=current.next
print("None")
```

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None

node1 =node(10) 
node2=node(20)  
node3=node(30)  
node4=node(40)   

node1.next=node2 #-->2000
node2.next=node3 #-->3000
node3.next=node4 #-->4000

head = node1
new_node= node(50)
new_node.next = head
head= new_node
current=head
while current is not None:
    print(current.data, end="->")
    current=current.next
print("None")
```

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None

node1 =node(10) 
node2=node(20)  
node3=node(30)  
node4=node(40)   

node1.next=node2 #-->2000
node2.next=node3 #-->3000
node3.next=node4 #-->4000


new_node= node(50)
head = node1
current=head
while current.next is not None:
    current=current.next
current.next= new_node
temp = head
while temp is not None:
    print(temp.data, end="->")
    temp=temp.next
print("None")
```
