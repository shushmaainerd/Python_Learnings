```python
result = ["heads","tails","tails","heads","tails","heads","heads","tails","tails","tails"]
heads_count=0
tails_count=0
for i in range(len(result)):
    print("Flip", i+1, "result", result[i])

    if result[i] == "heads":
        heads_count+=1
    elif result[i] == "tails":
        tails_count+=1 
print("heads count",heads_count)
print("tails count",tails_count)
```

```python
nums = [1, 2, 3, 4, 5, 6, 7]
even_count=0
odd_count=0
for i in nums:
    if i%2==0:
        even_count+=1
    else:
        odd_count+=1
print("even count", even_count)
print("odd count", odd_count)
```
